### Creation of Company Table

```python
Company_table=df[["Company","Company_Type","Industry"]].drop_duplicates().reset_index(drop=True)
Company_table
```

```python
Company_table.columns
```

```python
Company_table.insert(0, "Company_ID", ["C_" + str(i) for i in range(1, len(Company_table) + 1)])
```

```python
Company_table
```

### Creation of Location Table

```python
Location_table=df[["City","Location_Tier"]].drop_duplicates()
```

```python
Location_table.insert(0,"Location_ID",["L_"+ str(i) for i in range(1,len(Location_table)+1)])
```

```python
Location_table
```

```python
Fact_table=df.merge(Company_table,on=["Company", "Company_Type", "Industry"], how='left')
```

```python
Fact_table = Fact_table.merge(Location_table,on=["City", "Location_Tier"], how='left')
```

```python
Fact_table.drop(columns=["Company","Company_Type","Industry","City","Location_Tier"],inplace=True)
```

```python
Company_table.columns = [
    "company_id",
    "company",
    "company_type",
    "industry"
]

Location_table.columns = [
    "location_id",
    "city",
    "location_tier"
]

Fact_table.columns = [
    "job_id",
    "job_title",
    "experience_level",
    "job_type",
    "work_mode",
    "salary_lpa",
    "skills_required",
    "education_required",
    "openings",
    "applicants",
    "company_rating",
    "date_posted",
    "company_id",
    "location_id"
]
```

```python
print(Company_table.columns)
print(Location_table.columns)
print(Fact_table.columns)
```

```python
from sqlalchemy import text

create_tables = """

-- Drop existing tables
DROP TABLE IF EXISTS fact_table CASCADE;
DROP TABLE IF EXISTS company_table CASCADE;
DROP TABLE IF EXISTS location_table CASCADE;

----------------------------------------------------
-- Company Dimension
----------------------------------------------------

CREATE TABLE company_table
(
    company_id VARCHAR(20) PRIMARY KEY,
    company VARCHAR(255) NOT NULL,
    company_type VARCHAR(100),
    industry VARCHAR(100)
);

----------------------------------------------------
-- Location Dimension
----------------------------------------------------

CREATE TABLE location_table
(
    location_id VARCHAR(20) PRIMARY KEY,
    city VARCHAR(100),
    location_tier VARCHAR(50)
);

----------------------------------------------------
-- Fact Table
----------------------------------------------------

CREATE TABLE fact_table
(
    job_id VARCHAR(20) PRIMARY KEY,

    job_title VARCHAR(255),
    experience_level VARCHAR(100),
    job_type VARCHAR(50),
    work_mode VARCHAR(50),

    salary_lpa DECIMAL(10,2),

    skills_required TEXT,

    education_required VARCHAR(255),

    openings INT,

    applicants INT,

    company_rating DECIMAL(3,2),
    date_posted DATE,

    company_id VARCHAR(20),
    location_id VARCHAR(20),

    CONSTRAINT fk_company
        FOREIGN KEY(company_id)
        REFERENCES company_table(company_id),

    CONSTRAINT fk_location
        FOREIGN KEY(location_id)
        REFERENCES location_table(location_id)
);

"""

with engine.begin() as conn:
    conn.execute(text(create_tables))

print("Tables created successfully.")
```

```python
Fact_table.info()
```

```python
Company_table.to_sql(
    name="company_table",
    con=engine,
    if_exists="append",
    index=False
)

Location_table.to_sql(
    name="location_table",
    con=engine,
    if_exists="append",
    index=False
)

Fact_table.to_sql(
    name="fact_table",
    con=engine,
    if_exists="append",
    index=False
)

print("Data loaded successfully!")
```
